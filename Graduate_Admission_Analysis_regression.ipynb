```python
from google.colab import drive
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
# ML models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np
# Metrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
#Kfold
from sklearn.model_selection import KFold
# ML models
from sklearn.ensemble import RandomForestRegressor
# XGBoost
import xgboost as xgb
# CatBoost
from catboost import CatBoostRegressor
# LightGBM
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV
import numpy as np
# Metrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
#Kfold
from sklearn.model_selection import KFold

#----------------------Task 1 Data Pre-processing-------------------------------
data = pd.read_excel('/content/9_Graduate_Admission.xlsx')# 1) load and read data

print(data.shape)# 2) shape of data base (rows , columns)
print("\nNumber of variables (columns):", data.shape[1])
print("Number of instances (rows):", data.shape[0])

print(data.head())# 3) first few rows

summary = data.describe()# 4) explanatory data analysis to get descriptive statistics for all variables
print(summary)

# 5) separate dependent and independent variables
Y = data['Chance of Admit']#dependent variable (which we want to predict)
X = data.drop(['Chance of Admit'], axis=1)#independent which influence Y
#output shapes
# print("Shape of X (features):", X.shape)
# print("Shape of Y (target):", Y.shape)

# 6) Encode categorical variables and drop unnecessary columns
print(data['University Rating'].unique())#unique values in university rating
#Encode University Rating(Ordinal)
rating_mapping = {'Very poor':1 ,  'Poor':2 ,  'Average':3 , 'Very good':4 , 'Excellent':5 }
data['University Rating'] = data['University Rating'].map(rating_mapping)
#Encode Research(Binary)'
data['Research'] = data['Research'].astype(int)
X = data.drop(['Serial No.', 'Chance of Admit'], axis=1)#drop serial no


# 7) Check if any column has NaN values
print(data.isnull().sum()) #-> there is no null values

# 8) check dependence between variables
corr_matrix = data.corr()
# print(corr_matrix)#print correlation matrix
# #plot heatmap
# plt.figure(figsize=(10,8))
# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# plt.title("Correlation Heatmap")
# plt.show()

# 9) Scatter plots for each independent variable vs dependent variable
variables = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research']

# for var in variables:
#     plt.figure(figsize=(6,4))
#     sns.scatterplot(x=data[var], y=data['Chance of Admit'])
#     plt.title(f"{var} vs Chance of Admit")
#     plt.xlabel(var)
#     plt.ylabel("Chance of Admit")
#     plt.show()
#interpretation
'''The CGPA graphs show the strongest linear relationship with the admission probability, which means CGPA is likely the most influential feature for prediction.
GRE and TOEFL also show a linear relationship, but the points are more spread out (higher variance), so the relationship is weaker and less precise compared to CGPA.

For the University Rating, we see that higher ratings correspond to higher admission chances, with more observations concentrated at higher probabilities.
This indicates that university ranking has a noticeable impact on the outcome.

Regarding Research, it does not seem as important as the other features. The admission probability is already high even without research experience.
Research only appears to slightly extend the probability towards 100% in cases where it is positive.

For SOP and LOR, LOR appears slightly more precise, but overall both variables show weaker linear relationships and considerable spread.
This means they contribute to the prediction but are not as strong or reliable as CGPA or GRE.'''

#----------------------Task 2 Data Analysis-------------------------------
# 1) Split data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(#random_state to get same split every time
    X, Y, test_size=0.2, random_state=42
)
# print("X_train:", X_train.shape)
# print("X_test :", X_test.shape)
# print("Y_train:", Y_train.shape)
# print("Y_test :", Y_test.shape)

# 2) scale data with MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)#calculate max and min from taining set
#dont fit on test set to avoid data leakage
# Transform both training and test sets using max and min from traing set so the model cant learn about validation set
#x-xmin / xmax - xmin to transform each feature ti be between 0 and 1
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
#y is already between 0 and 1 so no need to scale
# 3) models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "SVM Regression": SVR(),
    "kNN Regression": KNeighborsRegressor()
}
#Train models and store metrics
r2_results = {}
mse_results = {}
mae_results = {}

for name, model in models.items():
    # Train model and output the regression score (R2), MSE, MAE of the model using training set
    model.fit(X_train_scaled, Y_train)
    y_train_pred = model.predict(X_train_scaled)
    r2_train = r2_score(Y_train, y_train_pred)
    mse_train = mean_squared_error(Y_train, y_train_pred)
    mae_train = mean_absolute_error(Y_train, y_train_pred)
    # Predictions test sets
    y_test_pred = model.predict(X_test_scaled)
    r2_test = r2_score(Y_test, y_test_pred)
    mse_test = mean_squared_error(Y_test, y_test_pred)
    mae_test = mean_absolute_error(Y_test, y_test_pred)

    r2_results[name] = [r2_train, r2_test]
    mse_results[name] = [mse_train, mse_test]
    mae_results[name] = [mae_train, mae_test]

r2_df = pd.DataFrame(r2_results, index=["Train", "Test"]).T
mse_df = pd.DataFrame(mse_results, index=["Train", "Test"]).T
mae_df = pd.DataFrame(mae_results, index=["Train", "Test"]).T

print("R2 scores:\n", r2_df)#the best model is the one with highest R2
print("\nMSE scores:\n", mse_df)#the best model is the one with lowest MSE
print("\nMAE scores:\n", mae_df)#the best model is the one with lowest MAE
#comparing results
''''By comparing the models, Linear Regression shows the highest R2 on the test set and the lowest MSE and MAE, which indicates strong generalization performance.
Both SVM and kNN provide reasonable results, but their test errors are still higher than those of Linear Regression, making Linear Regression the best-performing model
. In contrast, the Decision Tree model is clearly overfitting, as it achieves an R2 of 1 on the training set but significantly lower performance on the test set'''

# Plot squared error histograms (after all models trained)
# for name, model in models.items():
#     y_test_pred = model.predict(X_test_scaled)
#     squared_errors = (Y_test - y_test_pred)**2

#     plt.figure(figsize=(6,4))
#     plt.hist(squared_errors, bins=20, color='skyblue', edgecolor='black')
#     plt.title(f"{name} - Squared Error Histogram")
#     plt.xlabel("Squared Error")
#     plt.ylabel("Frequency")
#     plt.show()

# Interpretation of histograms of squared errors
'''The histograms of squared errors show that Linear Regression produces most of its errors between 0 and 0.02, indicating very stable and low-magnitude mistakes.
SVM shows a similar pattern but with a slightly larger tail reaching around 0.05.
kNN exhibits more frequent large errors, going up to approximately 0.07, suggesting less consistent performance.
Finally, Decision Tree has the widest spread and the largest errors, reaching around 0.12, which makes it the worst-performing model in terms of prediction accuracy.

These observations are fully consistent with the metric comparison'''

# 4) Hyperparameter tuning with GridSearchCV to tests combinations finds the best combination
param_grid_lr = {
    #try with and without intercept
    'fit_intercept': [True, False]
}
#train over 5 cross-validation
grid_lr = GridSearchCV(LinearRegression(), param_grid_lr, cv=5, scoring='r2')
grid_lr.fit(X_train_scaled, Y_train)
# print("Optimal parameters Linear Regression:", grid_lr.best_params_)

# Bar plot of coefficients
coefs = grid_lr.best_estimator_.coef_
features = X.columns
# plt.figure(figsize=(8,5))
# plt.bar(features, coefs, color='skyblue')
# plt.title("Linear Regression Coefficients")
# plt.xlabel("Features")
# plt.ylabel("Coefficient value")
# plt.show()
#interpretation
'''The bar plots of the regression coefficients perfectly match the interpretation we made from the independent dependent feature graphs.
CGPA has the strongest influence, followed by GRE and TOEFL, while SOP shows the smallest effect.
This confirms that CGPA is the most impactful predictor in the model,and SOP is the least important among the considered features.'''
#  Decision Tree Regressor
param_grid_dt = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'max_depth': [2, 4, 6, 8, None],
    'min_samples_split': [3, 5, 11],
    'min_samples_leaf': [1, 2, 4],
    'splitter': ['best', 'random'],
    'max_features': [None, 'sqrt', 'log2'],
    'max_leaf_nodes': [None, 10, 20, 50]

}
grid_dt = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid_dt, cv=5, scoring='r2')
grid_dt.fit(X_train_scaled, Y_train)
# print("Optimal parameters Decision Tree:", grid_dt.best_params_)

#  Support Vector Regressor
param_grid_svr = {
    'kernel': ['linear', 'rbf'],
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 0.2]
}
grid_svr = GridSearchCV(SVR(), param_grid_svr, cv=5, scoring='r2')
grid_svr.fit(X_train_scaled, Y_train)
# print("Optimal parameters SVR:", grid_svr.best_params_)

# knn Regressor
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
grid_knn = GridSearchCV(KNeighborsRegressor(), param_grid_knn, cv=5, scoring='r2')
grid_knn.fit(X_train_scaled, Y_train)
# print("Optimal parameters kNN:", gri d_knn.best_params_)

# 5) evaluate performance of best model:
best_models = {
    "Linear Regression": grid_lr.best_estimator_,
    "Decision Tree": grid_dt.best_estimator_,
    "SVM Regression": grid_svr.best_estimator_,
    "kNN Regression": grid_knn.best_estimator_
}

r2_best = {}
mse_best = {}
mae_best = {}

# Evaluate models on train and test sets
for name, model in best_models.items():
    y_train_pred = model.predict(X_train_scaled)
    y_test_pred = model.predict(X_test_scaled)

    r2_best[name] = [r2_score(Y_train, y_train_pred), r2_score(Y_test, y_test_pred)]
    mse_best[name] = [mean_squared_error(Y_train, y_train_pred), mean_squared_error(Y_test, y_test_pred)]
    mae_best[name] = [mean_absolute_error(Y_train, y_train_pred), mean_absolute_error(Y_test, y_test_pred)]

# Convert to DataFrames
r2_best_df = pd.DataFrame(r2_best, index=["Train", "Test"]).T
mse_best_df = pd.DataFrame(mse_best, index=["Train", "Test"]).T
mae_best_df = pd.DataFrame(mae_best, index=["Train", "Test"]).T

# Print results after optimization
# print("R2 scores (after optimization):")
# print(r2_best_df)
# print("\nMSE scores (after optimization):")
# print(mse_best_df)
# print("\nMAE scores (after optimization):")
# print(mae_best_df)
#compare results before and after optimization

#r2_df, mse_df, mae_df are from Task 2.3 (default parameters)
comparison_r2 = pd.concat([r2_df, r2_best_df], axis=1)
comparison_r2.columns = ['Train_Default', 'Test_Default', 'Train_Optimized', 'Test_Optimized']
# print("\nComparison of R2 (Default vs Optimized):")
# print(comparison_r2)

comparison_mse = pd.concat([mse_df, mse_best_df], axis=1)
comparison_mse.columns = ['Train_Default', 'Test_Default', 'Train_Optimized', 'Test_Optimized']
# print("\nComparison of MSE (Default vs Optimized):")
# print(comparison_mse)

comparison_mae = pd.concat([mae_df, mae_best_df], axis=1)
comparison_mae.columns = ['Train_Default', 'Test_Default', 'Train_Optimized', 'Test_Optimized']
# print("\nComparison of MAE (Default vs Optimized):")
# print(comparison_mae)
# Interpretation of optimization results
'''After hyperparameter optimization, the Decision Tree showed a significant improvement,
reducing overfitting and achieving better generalization. SVM and kNN saw slight performance gains, while Linear Regression remained largely unchanged,
indicating it was already stable. Overall, optimization either enhanced the model or maintained its performance.'''

# 6) create data frame to draw real results / prediction
#dataframe with 5 columns: "Real", "Linear Regression", "Decision Tree", "SVM Regression", "kNN Regression"
pred_df = pd.DataFrame()
pred_df['Real'] = Y_test.values  # Real values

for name, model in best_models.items():
    pred_df[name] = model.predict(X_test_scaled)

#real vs predicted values for 10 random test samples
sample_df = pred_df.sample(10, random_state=42)  # select 10 random rows
sample_df.plot(kind='bar', figsize=(10,5))
# plt.ylabel('Admission Chance')
# plt.title('Predictions vs Real Values for 10 Random Test Observations')
# plt.show()
#interpretation
'''Overall, kNN Regression provides the closest predictions, while Linear Regression and SVM also perform well but with slightly larger deviations.
Decision Tree shows more variability. For some test points (10, 4, 12), all models struggle, indicating that these observations are hard to predict,
possibly because they are outliers or not well represented in the training data. which is consistent with the results of Task 2.5
where these observations had the largest errors across all models.'''
#Parity plots for each model to visualize how close the predictions are to the real values
# for name in best_models.keys():
#     y_pred = pred_df[name]
#     plt.figure(figsize=(6,6))
#     plt.scatter(pred_df['Real'], y_pred, color='skyblue', edgecolor='black')
#     plt.plot([0,1],[0,1], 'r--')  # y=x line (perfect prediction)
#     plt.xlabel('Real Values')
#     plt.ylabel('Predicted Values')
#     plt.title(f'Parity Plot - {name}')
#     plt.show()
# interpretation
'''Linear Regression, kNN Regression, and SVM Regression The points lie close to the diagonal line, which indicates consistent and accurate predictions.
Decision Tree Regression The points are more scattered meaning the model’s predictions are less accurate and less consistent'''
# 7) Analyze prediction errors
errors_df = pd.DataFrame()
errors_df['Real'] = Y_test.values

for name, model in best_models.items():
    errors_df[name+'_error'] = Y_test.values - model.predict(X_test_scaled)
    errors_df[name+'_abs_error'] = np.abs(errors_df[name+'_error'])#absolute error

#Find biggest error for each model
for name in best_models.keys():
    idx_max = errors_df[name+'_abs_error'].idxmax()
    print(f"{name} - biggest error: {errors_df[name+'_error'].iloc[idx_max]:.4f} at observation index {idx_max}")

#Compare errors across models
errors_df['max_abs_error_all_models'] = errors_df[[name+'_abs_error' for name in best_models.keys()]].max(axis=1)#Max absolute error across all models for each observation

#Find observations badly predicted by all models
threshold = 0.1  # can adjust based on scale of your target
all_models_fail = errors_df[(errors_df[[name+'_abs_error' for name in best_models.keys()]] > threshold).all(axis=1)]
print("\nObservations badly predicted by ALL models:")
print(all_models_fail)
# 8)
random_states = [10, 21, 42, 56, 99]
r2_all, mse_all, mae_all = {name: [] for name in best_models.keys()}, {name: [] for name in best_models.keys()}, {name: [] for name in best_models.keys()}

for rs in random_states:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=rs)
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    for name, model in best_models.items():
        model.fit(X_train_scaled, Y_train)
        y_test_pred = model.predict(X_test_scaled)
        r2_all[name].append(r2_score(Y_test, y_test_pred))
        mse_all[name].append(mean_squared_error(Y_test, y_test_pred))
        mae_all[name].append(mean_absolute_error(Y_test, y_test_pred))

# Average metrics
r2_avg = {name: np.mean(v) for name,v in r2_all.items()}
mse_avg = {name: np.mean(v) for name,v in mse_all.items()}
mae_avg = {name: np.mean(v) for name,v in mae_all.items()}

print("Average R2:", r2_avg)
print("Average MSE:", mse_avg)
print("Average MAE:", mae_avg)
#interpretation of results across different random states
'''Based on the average metrics across multiple random train-test splits, Linear Regression emerges as the most reliable model.
It achieves the highest average R² (0.758) while maintaining the lowest average MSE (0.00498) and MAE (0.0494), indicating accurate and consistent predictions.
SVM Regression performs nearly as well, but slightly lower in R² and higher in errors. kNN Regression shows moderate performance,
and Decision Tree performs the worst, with lower R² and higher errors, suggesting poor generalization'''

# BoxPlots
# plt.figure(figsize=(12,4))
# plt.subplot(1,3,1); sns.boxplot(data=pd.DataFrame(r2_all)); plt.title("R2 BoxPlot"); plt.ylabel("R2")
# plt.subplot(1,3,2); sns.boxplot(data=pd.DataFrame(mse_all)); plt.title("MSE BoxPlot"); plt.ylabel("MSE")
# plt.subplot(1,3,3); sns.boxplot(data=pd.DataFrame(mae_all)); plt.title("MAE BoxPlot"); plt.ylabel("MAE")
# plt.tight_layout(); plt.show()
# Interpretation of stability analysis across different random states
'''Linear Regression is the best-performing and most reliable model.
It achieves the highest R2 score and the lowest MSE and MAE, showing that it predicts the data accurately. It also has the tightest box in the boxplots,
meaning its performance is the most stable and consistent across different data splits. SVM is a good alternative but still less accurate and less stable.
kNN and Decision Tree perform worse and show much larger variation.'''

# Bar plot of average metrics
metrics_avg_df = pd.DataFrame({'R2': r2_avg, 'MSE': mse_avg, 'MAE': mae_avg})
metrics_avg_df.plot(kind='bar', figsize=(10,5))
# plt.title("Average Performance Metrics Across 5 Random States")
# plt.ylabel("Metric Value")
# plt.show()
# Interpretation of bar plot
'''The results across the 5 random states confirm our previous findings: Linear Regression and SVM are the most accurate models,
while Decision Tree performs the worst.'''

#----------------------Task 3 Kfold cross validation --------------------------------

# 1) Set up K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # 5 folds, shuffle for randomness
# print("Indexes of observations in each fold:\n")
# for fold, (train_index, test_index) in enumerate(kf.split(X)):
#     print(f"Fold {fold+1}:")
#     print("Train indexes:", train_index)  # row numbers in training set
#     print("Test indexes:", test_index)    # row numbers in test set
#     print()

# 2) prepare the K-fold data
cv_results = {name: {'R2': [], 'MSE': [], 'MAE': []} for name in best_models.keys()}

for fold, (train_index, test_index) in enumerate(kf.split(X)):
    print(f"Processing Fold {fold+1}...")

    # Split the data into train/test for this fold
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    Y_train_fold, Y_test_fold = Y.iloc[train_index], Y.iloc[test_index]

    # Scale features using MinMaxScaler (fit on training fold only)
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train_fold)
    X_test_scaled = scaler.transform(X_test_fold)

#3) calculate  and evaluate models on each fold and calculate metrics
    # Train each model and evaluate metrics
    for name, model in best_models.items():
        # Train model on training fold
        model.fit(X_train_scaled, Y_train_fold)

        # Predict on test fold
        y_pred = model.predict(X_test_scaled)

        # Calculate metrics and store in dictionary
        cv_results[name]['R2'].append(r2_score(Y_test_fold, y_pred))
        cv_results[name]['MSE'].append(mean_squared_error(Y_test_fold, y_pred))
        cv_results[name]['MAE'].append(mean_absolute_error(Y_test_fold, y_pred))
#Output results per fold and compute average
print("\nR2 scores per fold and average:")
r2_df = pd.DataFrame({name: cv_results[name]['R2'] for name in cv_results.keys()},
                     index=[f"Fold {i+1}" for i in range(5)])
r2_df.loc['Average'] = r2_df.mean()
# print(r2_df)

print("\nMSE scores per fold and average:")
mse_df = pd.DataFrame({name: cv_results[name]['MSE'] for name in cv_results.keys()},
                      index=[f"Fold {i+1}" for i in range(5)])
mse_df.loc['Average'] = mse_df.mean()
# print(mse_df)

print("\nMAE scores per fold and average:")
mae_df = pd.DataFrame({name: cv_results[name]['MAE'] for name in cv_results.keys()},
                      index=[f"Fold {i+1}" for i in range(5)])
mae_df.loc['Average'] = mae_df.mean()
# print(mae_df)
avg_r2 = r2_df.loc['Average']
avg_mse = mse_df.loc['Average']
avg_mae = mae_df.loc['Average']

# 4)  conclusions about the best model and corresponding ML method.
'''Based on the evaluation across multiple metrics and folds, Linear Regression is the best model.
It consistently achieves the highest R² scores, the lowest MSE and MAE, and shows stable performance across different data splits.
SVM Regression is a close second but slightly less accurate and consistent. Decision Tree and kNN perform worse and are less reliable'''
# 5) Visualize average metrics across models
# Bar plot for R2
# plt.figure(figsize=(8,5))
# plt.bar(avg_r2.index, avg_r2.values, color='skyblue', edgecolor='black')
# plt.title("Average R2 per Model (5-Fold CV)")
# plt.ylabel("R2")
# plt.show()

# Bar plot for MSE
# plt.figure(figsize=(8,5))
# plt.bar(avg_mse.index, avg_mse.values, color='salmon', edgecolor='black')
# plt.title("Average MSE per Model (5-Fold CV)")
# plt.ylabel("MSE")
# plt.show()

# Bar plot for MAE
# plt.figure(figsize=(8,5))
# plt.bar(avg_mae.index, avg_mae.values, color='lightgreen', edgecolor='black')
# plt.title("Average MAE per Model (5-Fold CV)")
# plt.ylabel("MAE")
# plt.show()
#interpretation of visualizations
'''The 5 Fold Cross-Validation confirms the initial results from Task 2.8. Linear Regression consistently achieves the highest R2 and the lowest MSE and MAE,
making it the best-performing model. SVM Regression is a close second, while Decision Tree performs the worst.
This validates that Linear Regression’s superior performance is consistent and reliable across different data splits.'''
print("=========================================================================")
print("======================Assignmet 2===========================================")
print("=========================================================================")

X_train_2, X_test_2, Y_train_2, Y_test_2 = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

scaler_2 = MinMaxScaler()
X_train_scaled_2 = scaler_2.fit_transform(X_train_2)
X_test_scaled_2 = scaler_2.transform(X_test_2)


models_2 = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "XGBoost Regressor": xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    "CatBoost Regressor": CatBoostRegressor(iterations=100, learning_rate=0.1, verbose=0, random_state=42),
    "LightGBM Regressor": lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}


#Train models and store metrics
r2_results_2 = {}
mse_results_2 = {}
mae_results_2 = {}

for name, model in models_2.items():
    model.fit(X_train_scaled_2, Y_train_2)

    y_train_pred_2 = model.predict(X_train_scaled_2)
    y_test_pred_2 = model.predict(X_test_scaled_2)

    r2_train_2 = r2_score(Y_train_2, y_train_pred_2)
    mse_train_2 = mean_squared_error(Y_train_2, y_train_pred_2)
    mae_train_2 = mean_absolute_error(Y_train_2, y_train_pred_2)

    r2_test_2 = r2_score(Y_test_2, y_test_pred_2)
    mse_test_2 = mean_squared_error(Y_test_2, y_test_pred_2)
    mae_test_2 = mean_absolute_error(Y_test_2, y_test_pred_2)

    r2_results_2[name] = [r2_train_2, r2_test_2]
    mse_results_2[name] = [mse_train_2, mse_test_2]
    mae_results_2[name] = [mae_train_2, mae_test_2]

r2_df_2 = pd.DataFrame(r2_results_2, index=["Train", "Test"]).T
mse_df_2 = pd.DataFrame(mse_results_2, index=["Train", "Test"]).T
mae_df_2 = pd.DataFrame(mae_results_2, index=["Train", "Test"]).T

print("R2 scores:\n", r2_df_2)
print("\nMSE scores:\n", mse_df_2)
print("\nMAE scores:\n", mae_df_2)

# 2) Hyperparameter tuning with GridSearchCV to tests combinations finds the best combination
param_grid_rf = {
    # Random Forest Regressor
    'n_estimators': [50, 100, 200],
    'max_features': ['sqrt', 1.0],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

grid_rf = GridSearchCV(RandomForestRegressor(n_estimators=100, random_state=42), param_grid_rf, scoring='r2')
grid_rf.fit(X_train_scaled_2, Y_train_2)
print("Optimal parameters Random Forest:", grid_rf.best_params_)

param_grid_xgb = {
    # XGBoost Regressor
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'reg_alpha': [0, 0.1],
    'subsample': [0.8, 1.0]
}
grid_xgb = GridSearchCV(xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42, objective='reg:squarederror'), param_grid_xgb, cv=5, scoring='r2')
grid_xgb.fit(X_train_scaled_2, Y_train_2)
print("Optimal parameters XGBoost Regressor:", grid_xgb.best_params_)

param_grid_cb = {
    # CatBoost Regressor
    'depth': [4, 6, 8],
    'learning_rate': [0.03, 0.1],
    'l2_leaf_reg': [1, 3],
    'iterations': [50, 100]
}
grid_cb = GridSearchCV(CatBoostRegressor(iterations=100, learning_rate=0.1, verbose=0, random_state=42), param_grid_cb, cv=5, scoring='r2')
grid_cb.fit(X_train_scaled_2, Y_train_2)
print("Optimal parameters CatBoost Regressor:", grid_cb.best_params_)

param_grid_lgbm = {
    # LightGBM Regressor
    'max_depth': [5, 10],
    'num_leaves': [31, 60],
    'min_child_samples': [20, 50],
    'colsample_bytree': [0.8, 1.0]
}
grid_lgbm = GridSearchCV(lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42, verbose=-1), param_grid_lgbm, cv=5, scoring='r2')
grid_lgbm.fit(X_train_scaled_2, Y_train_2)
print("Optimal parameters LightGBM Regressor:", grid_lgbm.best_params_)

# 5) evaluate performance of best model:
best_models_2 = {
    "Random Forest": grid_rf.best_estimator_,
    "XGBoost Regressor": grid_xgb.best_estimator_,
    "CatBoost Regressor": grid_cb.best_estimator_,
    "LightGBM Regressor": grid_lgbm.best_estimator_
}


kf = KFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    X_train_fold, X_test_fold = X.iloc[train_idx], X.iloc[test_idx]
    y_train_fold, y_test_fold = Y.iloc[train_idx], Y.iloc[test_idx]

    # Scale
    scaler_fold = MinMaxScaler()
    X_train_scaled_fold = scaler_fold.fit_transform(X_train_fold)
    X_test_scaled_fold = scaler_fold.transform(X_test_fold)

    if fold == 0:
        fold_preds = {}
        for name, model in best_models_2.items():
            # Fit the model on fold training set
            model.fit(X_train_scaled_fold, y_train_fold)
            fold_preds[name + "_pred"] = model.predict(X_test_scaled_fold)

        # Create DataFrame: real values + predictions
        results_df = pd.DataFrame({'real': y_test_fold.values, **fold_preds})
        print("\nPredictions DataFrame (first fold):")
        print(results_df.head())
        # ======================= BAR PLOT FOR FOLD 0 =======================

        N = 10
        df_plot = results_df.head(N)

        x = np.arange(N)
        real = df_plot["real"]
        pred_cols = [c for c in df_plot.columns if c != "real"]

        plt.figure(figsize=(16, 6))
        width = 0.4 / (len(pred_cols) + 1)

        # Real values
        plt.bar(x, real, width, label="Real")

        # Predictions
        for i, col in enumerate(pred_cols):
            plt.bar(x + (i+1)*width, df_plot[col], width, label=col.replace("_pred", ""))

        # Use observation numbers as x-axis labels
        plt.xticks(x, [f"{i}" for i in range(1, N+1)])

        plt.xlabel("Observation Number")
        plt.ylabel("Value")
        plt.title("Real vs Predicted Values")
        plt.legend()
        plt.tight_layout()
        plt.show()


        # --- Parity plots ---
        plt.figure(figsize=(16, 12))
        for i, (name, y_pred) in enumerate(fold_preds.items(), 1):
            plt.subplot(2, 2, i)
            plt.scatter(y_test_fold, y_pred, alpha=0.6)
            plt.plot([y_test_fold.min(), y_test_fold.max()],
                     [y_test_fold.min(), y_test_fold.max()], 'r--')
            plt.xlabel("Actual")
            plt.ylabel("Predicted")
            plt.title(f"Parity plot: {name.replace('_pred','')}")

        plt.show()
        break  # only first fold

r2_best_2 = {}
mse_best_2 = {}
mae_best_2 = {}

# Evaluate models on train and test sets
for name, model in best_models_2.items():
    y_train_pred = model.predict(X_train_scaled_2)
    y_test_pred = model.predict(X_test_scaled_2)

    r2_best_2[name] = [r2_score(Y_train_2, y_train_pred), r2_score(Y_test_2, y_test_pred)]
    mse_best_2[name] = [mean_squared_error(Y_train_2, y_train_pred), mean_squared_error(Y_test_2, y_test_pred)]
    mae_best_2[name] = [mean_absolute_error(Y_train_2, y_train_pred), mean_absolute_error(Y_test_2, y_test_pred)]


r2_best_df_2 = pd.DataFrame(r2_best_2, index=["Train", "Test"]).T
mse_best_df_2 = pd.DataFrame(mse_best_2, index=["Train", "Test"]).T
mae_best_df_2 = pd.DataFrame(mae_best_2, index=["Train", "Test"]).T

r2_best_df_2.index = [name + " (Optimized)" for name in r2_best_df_2.index]
mse_best_df_2.index = [name + " (Optimized)" for name in mse_best_df_2.index]
mae_best_df_2.index = [name + " (Optimized)" for name in mae_best_df_2.index]


r2_model22 = pd.concat([r2_df_2, r2_best_df_2])
mse_df_22 = pd.concat([mse_df_2, mse_best_df_2])
mae_df_22 =pd.concat([mae_df_2, mae_best_df_2])


print("------------R2 scores (Train/Test) default / optimized ------------")
print(r2_model22)

print("\n------------MSE scores (Train/Test)  default / optimized  ------------")
print(mse_df_22)

print("\n------------MAE scores (Train/Test)  default / optimized  ------------")
print(mae_df_22)


r2_all = pd.concat([r2_best_df, r2_best_df_2])
mse_all = pd.concat([mse_best_df, mse_best_df_2])
mae_all = pd.concat([mae_best_df, mae_best_df_2])

# Optional: rename rows to indicate assignment
r2_all.index = [f"{name}" if name in r2_best_df.index else f"{name}" for name in r2_all.index]
mse_all.index = [f"{name}" if name in mse_best_df.index else f"={name}" for name in mse_all.index]
mae_all.index = [f"{name}" if name in mae_best_df.index else f"{name}" for name in mae_all.index]

# Display final DataFrames
print("------------R2 scores (Train/Test)------------")
print(r2_all)

print("\n------------MSE scores (Train/Test)------------")
print(mse_all)

print("\n------------MAE scores (Train/Test)------------")
print(mae_all)

top3_models = r2_all['Test'].sort_values(ascending=False).head(3)
print("Top 3 models based on Test R²:\n", top3_models)

plt.figure(figsize=(8,6))
plt.bar(top3_models.index, top3_models.values, color=['skyblue', 'lightgreen', 'salmon'])
plt.ylabel('R² (Test set)')
plt.title('Top 3 Models based on Test R²')
plt.ylim(0, 1)  # R² ranges from 0 to 1
plt.xticks(rotation=45, ha='right')
for i, v in enumerate(top3_models.values):
    plt.text(i, v + 0.02, f"{v:.3f}", ha='center', fontweight='bold')
plt.show()
```
