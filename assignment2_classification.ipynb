```python
!pip install catboost
from google.colab import drive
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
# ML models

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
# Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
#Kfold
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from catboost import CatBoostClassifier
import lightgbm as lgb

#----------------------Task 1 Data Pre-processing-------------------------------
data = pd.read_excel('/content/10_Student_Performance.xlsx')# 1) load and read data

print(data.shape)# 2) shape of data base (rows , columns)
print("\nNumber of variables (columns):", data.shape[1])
print("Number of instances (rows):", data.shape[0])

print(data.head())# 3) first few rows
summary = data.describe()# 4) explanatory data analysis to get descriptive statistics for all variables
print(summary)
# 5) separate dependent and independent variables
data['pass_fail'] = np.where(data['G3'] >= 10, 'pass', 'fail')
data['pass_fail'] = data['pass_fail'].map({'fail': 0, 'pass': 1})

categorical_cols = data.select_dtypes(include=['object']).columns
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True).astype(int)

# X = data.drop(['G3', 'pass_fail', ], axis=1)#independent which influence Y
X = data.drop(['G3', 'pass_fail', 'address_U' ,  'nursery_yes' , 'famsize_LE3' , 'Fedu' , 'Medu' , 'studytime' , 'Fjob_health' , 'Mjob_teacher' , 'Mjob_other' , 'Mjob_health' , 'health' , 'Dalc' , 'Pstatus_T' , 'guardian_other' , 'reason_other', 'Fjob_teacher' , 'reason_home' , 'activities_yes' , 'nursery_yes' , 'famsup_yes' , 'schoolsup_yes'], axis=1)#independent which influence Y
Y = data['pass_fail']##dependent variable (which we want to predict)
#output shapes
print("Shape of X (features):", X.shape)
print("Shape of Y (target):", Y.shape)

# 6) Encode categorical variables and drop unnecessary columns
# (already done above using get_dummies)
print(X.dtypes)
# 7) Check if any column has NaN values
print(data.isnull().sum()) #-> there is no null values
# 8) check dependence between variables
corr_matrix = data.corr()
print(corr_matrix)#print correlation matrix
#plot heatmap
plt.figure(figsize=(14, 14))  # Square figure
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            annot_kws={"size": 6},
            square=True)      # This makes cells square
plt.title("Correlation Heatmap", size=12)
plt.xticks(fontsize=7, rotation=90)
plt.yticks(fontsize=7)
plt.tight_layout()
plt.show()
# 9) Scatter plots for each independent variable vs dependent variable
for column in X.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=data, x=column, hue='pass_fail', bins=20, kde=False, palette=['red','green'], alpha=0.6)
    plt.title(f'Distribution of {column} by Pass/Fail')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
# 10) Determine number of classes in target variable
num_classes = Y.nunique()
print("Number of classes in the target variable:", num_classes)
print("Classes:", Y.unique())
#----------------------Task 2 Data Analysis-------------------------------
# 1) Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)

# Output shapes
print("Training set shape:", X_train.shape, Y_train.shape)
print("Testing set shape:", X_test.shape, Y_test.shape)
# 2) Feature Scaling using Min-Max Scaling
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# 3)
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'SVM': SVC(random_state=42, probability=True),
    'KNN': KNeighborsClassifier(n_neighbors=5)
}

acc_df = {}

for name, model in models.items():
    # Train Compute classification metrics for training set
    model.fit(X_train_scaled, Y_train)
    Y_train_pred = model.predict(X_train_scaled)  # Predictions for training set

    acc_train = accuracy_score(Y_train, Y_train_pred)

    # Train Compute classification metrics for testing set
    Y_test_pred = model.predict(X_test_scaled)  # Predictions for testing set

    acc_test  = accuracy_score(Y_test, Y_test_pred)

    # compare metrics obtained for training and testing sets
    acc_df[name] = [acc_train, acc_test]
    # Confusion matrix
    cm = confusion_matrix(Y_test, Y_test_pred)
    print(f"\n--- {name} ---")
    print("Confusion Matrix:\n", cm)
'''The results show that the Decision Tree overfits the training data (Train 1.0 vs Test 0.85),
while Logistic Regression and SVM achieve a good balance between training and test accuracy, with only slight overfitting.
KNN performs the worst overall, with both lower training and test accuracy. Overall, Logistic Regression and SVM are the most reliable models,
 whereas the Decision Tree tends to memorize the data, and KNN struggles to generalize.'''
# Output dataframes with metrics --------------------
acc_df = pd.DataFrame(acc_df, index=["Train", "Test"])
print("Accuracy Default :\n", acc_df)
'''Logistic Regression correctly detects most positives (102 TP) but struggles with negatives (9 TN, 11 FP).
Decision Tree is slightly more balanced (12 TN, 99 TP) but still misses some positives (11 FN). SVM and KNN are biased toward positives: SVM has 105 TP but only 3 TN,
while KNN has 102 TP and just 1 TN. Overall, Logistic Regression and Decision Tree handle both classes better, whereas SVM and KNN mostly fail on negatives.'''

for name, model in models.items():
    Y_test_pred = model.predict(X_test_scaled)
    cm = confusion_matrix(Y_test, Y_test_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()


best_model_acc = acc_df.loc["Test"].idxmax()
print(f"Best model by Accuracy on test set: {best_model_acc}")


# 4) Hyperparameter tuning with GridSearchCV to tests combinations finds the best combination
param_grid_logreg =  {
        'C': [0.01, 0.1, 1, 10, 100],
        'penalty': ['l1', 'l2'],
        'solver': ['saga'],  # saga supports l1, l2, elasticnet
        'max_iter': [1000]
    }
grid_logreg = GridSearchCV(LogisticRegression(random_state=42), param_grid_logreg, cv=5, scoring='accuracy')
grid_logreg.fit(X_train_scaled, Y_train)
print("Best hyperparameters for Logistic Regression:", grid_logreg.best_params_)
#Output and construct bar plots of the coefficients of the Logistic regression model
# Best Logistic Regression estimator
best_logreg = grid_logreg.best_estimator_
importance = best_logreg.coef_[0]

# Create a DataFrame
coef_df = pd.DataFrame({"Feature": X_train.columns, "Coefficient": importance})
coef_df["AbsCoeff"] = coef_df["Coefficient"].abs()#ignores the sign, only cares about strength of influence
coef_df = coef_df.sort_values(by="AbsCoeff", ascending=False)
print(coef_df)

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(x="Feature", y="Coefficient", data=coef_df)
plt.xticks(rotation=90)
plt.title("Logistic Regression Feature Coefficients")
plt.show()
'''The dominant features are G2 and G1, with high positive coefficients, strongly increasing the probability of a positive outcome.
The feature failures has a moderate negative impact, while features like Fjob_other, higher_yes, and goout have minor effects.
Most other features have near-zero coefficients, meaning Logistic Regression largely ignores them in predictions.'''

#decision tree
param_grid_dt = {
        'max_depth': [None, 3, 5, 7, 10],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'criterion': ['gini', 'entropy']
    }
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5, scoring='accuracy')
grid_dt.fit(X_train_scaled, Y_train)
print("Best hyperparameters for Decision Tree:", grid_dt.best_params_)
#SVM
param_grid_svm = {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': ['scale', 'auto']
    }
grid_svm = GridSearchCV(SVC(random_state=42), param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train_scaled, Y_train)
print("Best hyperparameters for SVM:", grid_svm.best_params_)
#KNN
param_grid_knn = {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan'],
        'p': [1, 2]  # 1=Manhattan, 2=Euclidean
    }
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, scoring='accuracy')
grid_knn.fit(X_train_scaled, Y_train)
print("Best hyperparameters for KNN:", grid_knn.best_params_)

# 5) evaluate tuned models and ourput results in the similar dataframe in default models
tuned_models = {
    'Logistic Regression': grid_logreg.best_estimator_,
    'Decision Tree': grid_dt.best_estimator_,
    'SVM': grid_svm.best_estimator_,
    'KNN': grid_knn.best_estimator_
}
acc_df_tuned = {}

for name, model in tuned_models.items():

    model.fit(X_train_scaled, Y_train)

    # Predictions for train and test
    Y_train_pred = model.predict(X_train_scaled)
    Y_test_pred = model.predict(X_test_scaled)

    # Train metrics
    acc_train = accuracy_score(Y_train, Y_train_pred)
    # Test metrics
    acc_test = accuracy_score(Y_test, Y_test_pred)

    # Store results
    acc_df_tuned[name] = [acc_train, acc_test]

# Convert dictionaries to DataFrames
acc_df_tuned = pd.DataFrame(acc_df_tuned, index=['Train', 'Test'])

print("\nTuned Model Accuracy:\n", acc_df_tuned)
acc_df.columns = acc_df_tuned.columns

# compare with default models
comparison_acc = acc_df_tuned - acc_df
print("\nAccuracy Improvement after Tuning:\n", comparison_acc)
'''Hyperparameter tuning improved performance for Logistic Regression, Decision Tree, and SVM, reducing overfitting and boosting test accuracy.
KNN did not benefit much, highlighting its sensitivity to data structure and scaling.'''
# 6)
# Create simple dataframe for real + predicted values
pred_df = pd.DataFrame()
pred_df['Real values'] = Y_test.values

for name, model in tuned_models.items():   # tuned_models
    pred_df[name] = model.predict(X_test_scaled)

# Select 10 random samples
sample_df = pred_df.sample(10, random_state=42)

# Simple bar plot
sample_df.plot(kind='bar', figsize=(12,6))
plt.ylabel('Predicted Class')
plt.title('Predictions vs Real Values for 10 Random Test Observations')
plt.xticks(rotation=0)
plt.show()
# Confusion matrix for each optimized model
for name, model in tuned_models.items():
    Y_test_pred = model.predict(X_test_scaled)
    cm = confusion_matrix(Y_test, Y_test_pred)

    print(f"\nConfusion Matrix - {name}:\n")
    print(cm)

    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
'''Logistic Regression, Decision Tree, and SVM all improved significantly, especially in reducing false positives (better TN) and balancing predictions.
KNN improved only slightly, still misclassifying many negatives'''
# 7) Compute prediction errors for each test observation for each tuned model
print("\n")
print("\n")
error_df = pd.DataFrame()
error_df["Real"] = Y_test.values

# Store predictions again (for clarity)
for name, model in tuned_models.items():
    error_df[f"Pred_{name}"] = model.predict(X_test_scaled)

# Compute errors = 0 (correct) or 1 (incorrect)
for name in tuned_models.keys():
    error_df[f"Error_{name}"] = (error_df["Real"] != error_df[f"Pred_{name}"]).astype(int)
print(error_df.head())
# Count errors per model
model_errors = {name: error_df[f"Error_{name}"].sum() for name in tuned_models.keys()}

print("\nTotal Misclassifications per Model:")
for model, err in model_errors.items():
    print(f"{model}: {err}")

# Identify observations misclassified by ALL models
error_df["All_Wrong"] = error_df[[f"Error_{name}" for name in tuned_models.keys()]].sum(axis=1)
badly_predicted_by_all = error_df[error_df["All_Wrong"] == len(tuned_models)]

print("\nObservations badly predicted by ALL models:")
print(badly_predicted_by_all)

# Plot error counts
plt.figure(figsize=(8,5))
plt.bar(model_errors.keys(), model_errors.values())
plt.title("Number of Misclassified Observations per Model")
plt.ylabel("Count of Errors")
plt.xticks(rotation=15)
plt.show()
#8) train using different random_state

random_states = [0, 21, 42, 77, 123]
performance_results = []

for rs in random_states:

    # New split
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=rs
    )

    # Scale again
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train all tuned models again
    for name, model in tuned_models.items():
        model.fit(X_train_scaled, Y_train)
        Y_pred = model.predict(X_test_scaled)

        performance_results.append({
            "Model": name,
            "RandomState": rs,
            "Accuracy": accuracy_score(Y_test, Y_pred),
        })

# Convert to DataFrame
results_df = pd.DataFrame(performance_results)
# Accuracy table
accuracy_table = results_df.pivot(index="RandomState", columns="Model", values="Accuracy")
print("Accuracy Table for Each Random Split:\n")
print(accuracy_table)
#plot boxplota
metrics = ["Accuracy"]
for metric in metrics:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=results_df, x="Model", y=metric)
    plt.title(f"Boxplot of {metric} Across 5 Random States")
    plt.xticks(rotation=15)
    plt.show()

avg_results = results_df.groupby("Model")[metrics].mean()
# Plot averages
for metric in metrics:
    plt.figure(figsize=(7,4))
    avg_results[metric].plot(kind="bar")
    plt.title(f"Average {metric} Across 5 Random States")
    plt.ylabel(metric)
    plt.xticks(rotation=15)
    plt.show()
'''Logistic Regression and SVM consistently perform best, showing stability across splits.
Decision Tree has slightly higher peaks but can fluctuate depending on data, and KNN struggles to generalize.'''
#--------------------------------Task 3 : Kfold cross validation -------------------------\
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("K-Fold(indexes of observations)")
fold_number = 1
for train_index, test_index in kf.split(X):
    print(f"\nFold {fold_number}:")
    print("Train indexes:", train_index)
    print("Test indexes:", test_index)
    fold_number += 1

# 2) training models using K fold
kfold_results = []   # store all accuracies

for name, model in tuned_models.items():
    print(f"\n Model: {name} : ")
    fold = 1

    for train_index, test_index in kf.split(X):

        # Split data
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]

        # Scale features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # Train model
        model.fit(X_train_scaled, Y_train)

        # Predict
        Y_pred = model.predict(X_test_scaled)

        # Compute accuracy
        acc = accuracy_score(Y_test, Y_pred)

        # Save result
        kfold_results.append({
            "Model": name,
            "Fold": fold,
            "Accuracy": acc
        })

        fold += 1
print("-----------------------------------------------------")
# 3)
kfold_df = pd.DataFrame(kfold_results)

print("\nAccuracy per Fold")
print(kfold_df)

# average accuracy for each model
avg_accuracy = kfold_df.groupby("Model")["Accuracy"].mean()

print("\nAverage Accuracy per Model")
print(avg_accuracy)
# 4)
'''After K-Fold cross-validation, Logistic Regression achieves the highest average accuracy (0.924),
followed by SVM (0.908) and Decision Tree (0.900), while KNN performs worst (0.871). This confirms that Logistic Regression is the most reliable model for this dataset,
with KNN being the least consistent'''
# 5) Bar Plot of Average Performance per Model

plt.figure(figsize=(7,5))
avg_accuracy.plot(kind="bar")
plt.title("Average Accuracy of Each Model (5-Fold Cross Validation)")
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.show()


print("=========================================================================")
print("======================Assignmet 2===========================================")
print("=========================================================================")


# ----------------- 1) Train-Test Split + Scaling -----------------
X_train_2, X_test_2, Y_train_2, Y_test_2 = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

scaler_2 = MinMaxScaler()
X_train_scaled_2 = scaler_2.fit_transform(X_train_2)
X_test_scaled_2 = scaler_2.transform(X_test_2)

# ----------------- 2) Default Models -----------------
models_2 = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, use_label_encoder=False, eval_metric='logloss', random_state=42),
    "CatBoost": CatBoostClassifier(iterations=100, learning_rate=0.1, verbose=0, random_state=42),
    "LightGBM": lgb.LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
}


# ----------------- 3) Train with Default Hyperparameters -----------------
acc_default = {}
conf_matrices_default = {}

for name, model in models_2.items():
    # Train
    model.fit(X_train_scaled_2, Y_train_2)
    y_train_pred = model.predict(X_train_scaled_2)
    y_test_pred = model.predict(X_test_scaled_2)

    # Accuracy
    acc_train = accuracy_score(Y_train_2, y_train_pred)
    acc_test  = accuracy_score(Y_test_2, y_test_pred)
    acc_default[name] = [acc_train, acc_test]

    # Confusion matrix
    conf_matrices_default[name] = confusion_matrix(Y_test_2, y_test_pred)
    print(f"\n--- {name} --- Default ---")
    print("Confusion Matrix:\n", conf_matrices_default[name])

# Heatmaps for default models
for name, cm in conf_matrices_default.items():
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name} (Default)')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

acc_df_default = pd.DataFrame(acc_default, index=["Train Default", "Test Default"]).T
print("\nAccuracy (Default Hyperparameters):\n", acc_df_default)

# ----------------- 4) Grid Search for Optimized Hyperparameters -----------------
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_features': ['sqrt', 1.0],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

param_grid_xgb = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'reg_alpha': [0, 0.1],
    'subsample': [0.8, 1.0]
}

param_grid_cb = {
    'depth': [4, 6, 8],
    'learning_rate': [0.03, 0.1],
    'l2_leaf_reg': [1, 3],
    'iterations': [50, 100],

}

param_grid_lgbm = {
    'max_depth': [5, 10, -1],
    'num_leaves': [31, 60],
    'min_child_samples': [20, 50],
    'colsample_bytree': [0.8, 1.0]
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42),
                       param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1, refit=True)
grid_xgb = GridSearchCV(xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
                        param_grid_xgb, cv=5, scoring='accuracy', n_jobs=-1, refit=True)
grid_cb = GridSearchCV(CatBoostClassifier(verbose=0, random_state=42),
                       param_grid_cb, cv=5, scoring='accuracy', n_jobs=-1, refit=True)
grid_lgbm = GridSearchCV(lgb.LGBMClassifier(random_state=42),
                         param_grid_lgbm, cv=5, scoring='accuracy', n_jobs=-1, refit=True)

# Fit grids
grid_rf.fit(X_train_scaled_2, Y_train_2)
print("Best hyperparameters for Random Forest:", grid_rf.best_params_)

grid_xgb.fit(X_train_scaled_2, Y_train_2)
print("Best hyperparameters for XGBoost:", grid_xgb.best_params_)

grid_cb.fit(X_train_scaled_2, Y_train_2)
print("Best hyperparameters for CatBoost:", grid_cb.best_params_)

grid_lgbm.fit(X_train_scaled_2, Y_train_2)
print("Best hyperparameters for LightGBM:", grid_lgbm.best_params_)


# best_models_2 = {
#     "Random Forest": grid_rf.best_estimator_,
#     "XGBoost": grid_xgb.best_estimator_,
#     "CatBoost": models_2["CatBoost"],
#     "LightGBM": grid_lgbm.best_estimator_
# }
best_models_2 = {
    "Random Forest": grid_rf.best_estimator_,
    "XGBoost": grid_xgb.best_estimator_,
    "CatBoost": models_2["CatBoost"],  # already trained default instance
    "LightGBM": grid_lgbm.best_estimator_
}
# ----------------- 3) Prediction Results for Test Set (First CV Fold) -----------------
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    X_train_fold, X_test_fold = X.iloc[train_idx], X.iloc[test_idx]
    y_train_fold, y_test_fold = Y.iloc[train_idx], Y.iloc[test_idx]

    # Scale
    scaler_fold = MinMaxScaler()
    X_train_scaled_fold = scaler_fold.fit_transform(X_train_fold)
    X_test_scaled_fold = scaler_fold.transform(X_test_fold)

    if fold == 0:
        # Create DataFrame with real + predictions
        fold_preds = {}
        for name, model in best_models_2.items():
            model.fit(X_train_scaled_fold, y_train_fold)
            fold_preds[name + "_pred"] = model.predict(X_test_scaled_fold)

        results_df = pd.DataFrame({'real': y_test_fold.values, **fold_preds})
        print("\nPredictions DataFrame (first fold):")
        print(results_df)
        break  # Only first fold
# ----------------- 5) Train/Test Metrics for Optimized Models -----------------
#accuracy from the previous assignment

acc_df_tuned = pd.DataFrame(acc_df_tuned, index=['Train', 'Test'])

acc_optimized_2 = {}
conf_matrices_optimized = {}

for name, model in best_models_2.items():
    y_train_pred = model.predict(X_train_scaled_2)
    y_test_pred  = model.predict(X_test_scaled_2)

    acc_train = accuracy_score(Y_train_2, y_train_pred)
    acc_test  = accuracy_score(Y_test_2, y_test_pred)
    acc_optimized_2[name] = [acc_train, acc_test]

acc_df_optimized = pd.DataFrame(acc_optimized_2, index=["Train Optimized", "Test Optimized"]).T
print("\nAccuracy (Optimized Hyperparameters):\n", acc_df_optimized)


acc_df_default = acc_df_default.rename(columns={'Train': 'Train Default', 'Test': 'Test Default'})
acc_df_optimized = acc_df_optimized.rename(columns={'Train': 'Train Optimized', 'Test': 'Test Optimized'})
# Concatenate along columns
combined_df = pd.concat([acc_df_default, acc_df_optimized], axis=1)
# Reorder columns for readability
combined_df = combined_df[['Train Default', 'Test Default', 'Train Optimized', 'Test Optimized']]
print("\nAccuracy (Default & Optimized Hyperparameters):\n", combined_df)



for name, model in best_models_2.items():
    conf_matrices_optimized[name] = confusion_matrix(y_test_fold, results_df[name+'_pred'])

for name, cm in conf_matrices_optimized.items():
    print(f"\nConfusion Matrix: {name}\n{cm}")

# Heatmaps for optimized models
for name, cm in conf_matrices_optimized.items():
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name} (Optimized)')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()


# Transpose so rows = models, columns = Train/Test
acc_df_tuned_fixed = acc_df_tuned.T.copy()
acc_df_tuned_fixed.index.name = None
acc_df_tuned_fixed.columns = ['Train', 'Test']  # rename for consistency
acc_df_tuned_fixed.index = [f"Assignment1_{name}" for name in acc_df_tuned_fixed.index]

acc_df_optimized_fixed = acc_df_optimized.copy()
acc_df_optimized_fixed.columns = ['Train', 'Test']
acc_df_optimized_fixed.index = [f"Assignment2_{name}" for name in acc_df_optimized_fixed.index]

acc_all = pd.concat([acc_df_tuned_fixed, acc_df_optimized_fixed])
print("\nCombined Accuracy (Train/Test) for all models:\n", acc_all)

top3_acc_models = acc_all['Test'].sort_values(ascending=False).head(3)
print("\nTop 3 models based on Test Accuracy:\n", top3_acc_models)

plt.figure(figsize=(8,6))
plt.bar(top3_acc_models.index, top3_acc_models.values, color=['skyblue', 'lightgreen', 'salmon'])
plt.ylabel('Accuracy (Test set)')
plt.title('Top 3 Models based on Test Accuracy')
plt.ylim(0, 1)
for i, v in enumerate(top3_acc_models.values):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center', fontweight='bold')
plt.show()
```
